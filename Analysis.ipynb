```python
#lets see Transaction Status Counts of success or Failed
df['transaction_status'].value_counts(normalize=True) * 100 #for %
```

```python
#lets in which hour the trafic is lower and higher
df['hour_of_day'].value_counts().sort_values(ascending=False)
```

```python
# evening 7:00 pm the trafic is very high and 3:00 am is time where is less trafic
# lets plot a graph by the help of gpt
df['hour_of_day'].value_counts().sort_index().plot(kind='bar',title='Transactions by hour')
```

```python
#lets see whats going on the weekends
df.groupby('is_weekend')['amount (INR)'].agg(['count','sum','mean'])
```

```python
# all right lets do some Age group related analysis
#top Sender's
df['sender_age_group'].value_counts().head(10)
```

```python
#Fraud rate by sender age group
df.groupby('sender_age_group')['fraud_flag'].mean().sort_values(ascending=False)
```

```python
#failure rate per banks
round(df.groupby('sender_bank')['transaction_status'].value_counts(normalize=True).unstack().fillna(0) *100,2)
```

```python
#Top Banks used for sending money
df['sender_bank'].value_counts()
```

```python
#Now lets check device fraud rate
df.groupby('device_type')['fraud_flag'].mean().sort_values(ascending=False)
```

```python
#That surpise me Web have the most rate 
#Now lets see a Category analysis
df.groupby('merchant_category')['amount (INR)'].sum().sort_values(ascending=False)
```

```python
#lets Check in which hour Fraud are done 
df[df['fraud_flag']==1]['hour_of_day'].value_counts().sort_index()
```

```python
#lets plot a graph by the help of gpt
df[df['fraud_flag']==1]['hour_of_day'].value_counts().sort_index().plot(kind='bar',title='Fraud Transactions by hour')
```

```python
#fraud rate by Transaction type 
pd.crosstab(df['transaction type'],df['fraud_flag'],normalize='index')*100
```

```python
#and by network
pd.crosstab(df['network_type'],df['fraud_flag'],normalize='index')*100
```
